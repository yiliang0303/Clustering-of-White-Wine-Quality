```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_path = ['data']
traindir = ['train']
testdir = ['test']
```

### Original Dataset (multiclass target)

```python
trainpath = os.sep.join(data_path + traindir + ['Train.csv'])

traindf = pd.read_csv(trainpath, sep = ",")
traindf
```

```python
testpath = os.sep.join(data_path + testdir + ['Test.csv'])

testdf = pd.read_csv(testpath, sep = ",")
testdf
```

```python
X_train = traindf.drop("quality", axis=1)
X_test = testdf.drop("quality", axis=1)
y_train = traindf["quality"]
y_test = testdf["quality"]
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

### Modified Dataset (binary target)

```python
trainpath = os.sep.join(data_path + traindir + ['Train_Modified.csv'])

traindf = pd.read_csv(trainpath, sep = ",")
traindf
```

```python
testpath = os.sep.join(data_path + testdir + ['Test_Modified.csv'])

testdf = pd.read_csv(testpath, sep = ",")
testdf
```

```python
X_train = traindf.drop("quality", axis=1)
X_test = testdf.drop("quality", axis=1)
y_train = traindf["quality"]
y_test = testdf["quality"]
```

```python
#Import svm model
from sklearn.svm import SVC

#Create a svm Classifier
clf = SVC(kernel='rbf') # RBF Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# print prediction results
print(classification_report(y_test, y_pred, digits = 4))
```

Standard Scaler

```python
# Creating scaled set to be used in model to improve the results
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Create a svm Classifier
clf = SVC(kernel='rbf') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# print prediction results
print(classification_report(y_test, y_pred, digits = 4))
```

```python
# SVM Hyperparameter Tuning using GridSearchCV | ML
# https://www.geeksforgeeks.org/svm-hyperparameter-tuning-using-gridsearchcv-ml/
```

```python
from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
  
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)
  
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
# print best parameter after tuning
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
grid_predictions = grid.predict(X_test)
  
# print classification report
print(classification_report(y_test, grid_predictions, digits = 4))
```

**86.87%** Prediction Result, an improvement of 9.25%
