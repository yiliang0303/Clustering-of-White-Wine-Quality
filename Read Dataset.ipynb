## Dataset Preprocessing
Since Quality (Target Variables) has outcomes from 3 to 9 <br>
We modify the Quality Columns to Binary Outcomes <br>
If Quality is 7 or more , it's **Good** Quality. <br>
If Quality less than 7 , it's **Poor** Quality. <br>

```python
data['quality'] = np.where((data.quality >= 7),'Good',"Poor").astype(str)
data
```

```python
filepath = os.sep.join(data_path + ['winequality-white_modifed.csv'])
data.to_csv(filepath, index = False)
```
